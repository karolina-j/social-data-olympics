## Preprocessing

1. Fiter the Olympic results:
    - Keep only rows where a medal was awarded
    - Drop rows with missing 'year' ==> this is only the case for the "Intercalated Games" from 1906
    - Drop "youth" events
    - Keep only Olympics from 1972 onward

```python
# Only rows where a medal was awarded
medals = results[results["medal"].isin(["Gold", "Silver", "Bronze"])].copy()

# Drop rows with missing 'year'
medals = medals.dropna(subset=["year"])

# Drop "youth" events
medals = medals[~medals['event'].str.contains('youth', case=False, na=False)]

# Drop events with (YOG) in the name
medals = medals[~medals['event'].str.contains('YOG', case=False, na=False)]

# Only Olympics from 1972 onward
medals = medals[medals["year"] >= 1972]
```

2. Adjust the 'year' column by converting it to integers for consistency

```python
medals["year"] = medals["year"].astype(int)
```

3. Map NOC codes to country names

```python
# NOC to country
medals = medals.merge(noc_map[["NOC", "region"]], left_on="noc", right_on="NOC", how="left")
medals = medals.rename(columns={"region": "country"})
```

4. Drop duplicates to avoid overcounting team medals

```python
# drop duplicates of team medals
dedup_medals = medals.drop_duplicates(
    subset=["year", "country", "event", "medal"],  # defines a unique medal award
    keep="first"
)
```

5. Make the medal table more readable
    - drop unwanted columns
    - reorder columns

```python
# Drop unwanted columns
medals = medals.drop(columns=['NOC'])

# reorder columns
medals = medals[["year", "type", "medal", "country", "noc", "discipline", "event", "as", "athlete_id", "team", "place", "tied"]]
```

```python
medals.head()
```

## Data prep

### Create a Medal Counts Table

In this section, we create a medal counts table that will have the following columns:

| **Year** | **Country** | **Bronze** | **Silver** | **Gold** | **Total** |
|----------|-------------|------------|------------|----------|-----------|

This table will summarize the medal count for each country by year, providing a clear overview of Olympic performance.


```python
# Count medals by year, country, and medal type
counts = (
    dedup_medals
    .groupby(["year", "type", "country", "medal"])
    .size()
    .unstack(fill_value=0)
    .reset_index()
)
counts.columns.name = None  # This removes the name of the columns index
counts = counts[["year","type", "country", "Bronze", "Silver", "Gold"]]


# Ensure all medal columns exist, then calculate totals
for m in ["Gold", "Silver", "Bronze"]:
    if m not in counts.columns:
        counts[m] = 0
counts["Total"] = counts[["Gold", "Silver", "Bronze"]].sum(axis=1)
# counts.head()
print(counts)
counts.to_csv("medal_counts.csv", index=False)
```

# Population

## Preprocessing

```python
population = pd.read_csv("population.csv")
print(population.head())
```

Clean the Population Dataset

- Filter population data from 1972 and above




```python
population_1972_above = population[population["Year"] >= 1972]
population_1972_above = population_1972_above.rename(columns={"Population (historical)": "Population"})
population_1972_above['Entity'] = population_1972_above['Entity'].replace({
    'Czechia': 'Czech Republic',
    'United States': 'USA',
    'United Kingdom': 'UK',
    'Trinidad and Tobago': 'Trinidad',
    'United States Virgin Islands': 'Virgin Islands, US',
    'North Macedonia': 'Macedonia',
})
print(population_1972_above.head())
```

Merge the Population Dataset with Medals

```python
# Merge population data with counts dataframe
counts = counts.merge(
    population_1972_above[['Year', 'Entity', 'Population']],
    left_on=['year', 'country'],
    right_on=['Year', 'Entity'],
    how='left'
)

# Drop redundant columns after merging
counts = counts.drop(columns=['Year', 'Entity'])
counts.head()
```

## Visualizations

#### Olympic Medal Distribution Worldwide from 1972 to 2022

```python
import plotly.express as px

medal_totals = counts.groupby('country')['Total'].sum().reset_index()

fig = px.choropleth(
    medal_totals,
    locations="country",
    locationmode="country names",
    color="Total",
    hover_name="country",
    hover_data={"country": False},
    color_continuous_scale=px.colors.sequential.Reds,
    projection="orthographic",
    labels={"Total": "Total Medals"}
)

fig.update_traces(marker_line_width=0.5, marker_line_color='black')

fig.update_layout(
    title={
        'text': "Olympic Medal Distribution Worldwide from 1972 to 2022",
        'x': 0,
        'xanchor': 'left'
    },
    autosize=True,
    height=600,
    margin=dict(t=50, b=0, l=0, r=0)
)

fig.show()
```

# AHDI

Augmented HDI dataset was created by the Rafael del Pino Foundation <br>

link: https://frdelpino.es/investigacion/en_gb/world-economy/human-development-world-economy/

## Preprocessing


1. Import

```python
import pandas as pd

df = pd.read_csv('AHDI_Table8.csv')

# new list of column names
new_cols = []
for col in df.columns:
    try:
        year = int(float(col))
        new_cols.append(str(year))
    except ValueError:
        new_cols.append(col)

df.columns = new_cols
```

2. Filter dataset

```python
# filter years ≥ 1970
year_cols = [c for c in df.columns if c.isdigit() and int(c) >= 1970]

# keep only AHDI + year columns
df_1970_on = df[['AHDI'] + year_cols]
```

3. Create AHDI table

Columns: country,	year,	AHDI

```python
# wide to long format
df_ahdi = df_1970_on.melt(id_vars='AHDI', var_name='year', value_name='AHDI_Value')

# rename columns for clarity
df_ahdi = df_ahdi.rename(columns={'AHDI': 'country', 'AHDI_Value': 'AHDI'})

df_ahdi['year'] = df_ahdi['year'].astype(int)

# standardize country names in the HDI dataset
df_ahdi['country'] = df_ahdi['country'].replace({
    "Korea, South": "South Korea",
    "Brunei Darussalam": "Brunei"
})

# placeholder rows for missing HDI countries
missing_hdi_countries = pd.DataFrame({
    'country': ['San Marino', 'Liechtenstein'],
    'AHDI': [None, None],
    'year': [None, None]
})

# append the placeholder rows
df_ahdi = pd.concat([df_ahdi, missing_hdi_countries], ignore_index=True)
df_ahdi['year'] = df_ahdi['year'].astype('Int64')

# reshaped DataFrame
df_ahdi.head()
```

Define color palette

```python
grey = "rgba(235,235,235,0.5)"

chartBlue = "#68BEF0"
chartYellow = "#FFBD4C"
chartGreen = "#008B44"
```

## Visualizations

#### Average medal count vs average AHDI for all countries over the entire period


```python
# aggregate total medals by country
total_medals_by_country = counts.groupby('country', as_index=False)['Total'].sum()

# filter HDI data for years ≥ 1972
filtered_hdi = df_ahdi[df_ahdi['year'] >= 1972]

# aggregate average HDI by country
avg_hdi_by_country = filtered_hdi.groupby('country', as_index=False)['AHDI'].mean()

# merge the two datasets on 'country'
merged_data = pd.merge(total_medals_by_country, avg_hdi_by_country, on='country')

# determine AHDI brackets dynamically using quantiles
hdi_quantiles = merged_data['AHDI'].quantile([0.25, 0.5, 0.75])

label1 = f'Low (AHDI < {round(hdi_quantiles[0.25], 2)})'
label2 = f'Medium ({round(hdi_quantiles[0.25], 2)} - {round(hdi_quantiles[0.5], 2)})'
label3 = f'High ({round(hdi_quantiles[0.5], 2)} - {round(hdi_quantiles[0.75], 2)})'
label4 = f'Very High (AHDI > {round(hdi_quantiles[0.75], 2)})'

def categorize_hdi(hdi):
    if hdi < hdi_quantiles[0.25]:
        return label1
    elif hdi_quantiles[0.25] <= hdi < hdi_quantiles[0.5]:
        return label2
    elif hdi_quantiles[0.5] <= hdi < hdi_quantiles[0.75]:
        return label3
    else:
        return label4

# categorize countries into AHDI brackets
merged_data['AHDI_Bracket'] = merged_data['AHDI'].apply(categorize_hdi)

# aggregate total medals by AHDI bracket
medals_by_hdi_bracket = merged_data.groupby('AHDI_Bracket', as_index=False)['Total'].sum()

# define the correct order based on descriptive labels
custom_order = [label1, label2, label3, label4]

# custom categorical order
medals_by_hdi_bracket['AHDI_Bracket'] = pd.Categorical(
    medals_by_hdi_bracket['AHDI_Bracket'],
    categories=custom_order,
    ordered=True
)
medals_by_hdi_bracket.sort_values('AHDI_Bracket', inplace=True)
```

```python
fig = px.bar(
    medals_by_hdi_bracket,
    x='AHDI_Bracket',
    y='Total',
    text=None,
    color='AHDI_Bracket',
    labels={'AHDI_Bracket': 'Augmented HDI (AHDI)', 'Total': 'Total Medal Count'},
    title='Olympic Medal Count by Augmented HDI Bracket (1972 - 2022)'
)

fig.update_traces(
    hovertemplate='HDI Bracket: %{x}<br>Total Medals: %{y} <extra></extra>',
    marker_color=chartGreen,
    showlegend=False,
)

fig.update_layout(
    xaxis_title='Augmented HDI (AHDI) Bracket',
    yaxis_title='Total Medal Count',
    template='plotly_white',
    legend=dict(
        x=1.05,
        y=0.5,
        bgcolor=grey,
        bordercolor='black',
        borderwidth=1
    ),
    margin=dict(l=50, r=150, t=50, b=100),
    title=dict(
        text='Olympic Medal Count by Augmented HDI Bracket (1972 - 2022)',
        font=dict(size=18)
    )
)

fig.show()
```
